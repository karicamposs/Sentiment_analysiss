**Data preparation**

```python
import re
import unicodedata
from nltk.corpus import stopwords

def clean_tokenize(texto):
    # Se convierte todo el texto a minúsculas
    nuevo_texto = texto.lower()
    
    # Eliminación de acentos y caracteres especiales
    nuevo_texto = unicodedata.normalize('NFKD', nuevo_texto).encode('ascii', 'ignore').decode('ascii')

    # Eliminación de páginas web (palabras que empiezan por "http")
    nuevo_texto = re.sub('http\S+', ' ', nuevo_texto)

    # Eliminación de signos de puntuación
    regex = '[\\!\\"\\#\\$\\%\\&\\\'\\(\\)\\*\\+\\,\\-\\.\\/\\:\\;\\<\\=\\>\\?\\@\\[\\\\\\]\\^_\\`\\{\\|\\}\\~]'
    nuevo_texto = re.sub(regex , ' ', nuevo_texto)

    # Eliminación de números
    nuevo_texto = re.sub("\d+", ' ', nuevo_texto)

    # Eliminación de espacios en blanco múltiples
    nuevo_texto = re.sub("\\s+", ' ', nuevo_texto)

    # Tokenización por palabras individuales
    nuevo_texto = nuevo_texto.split(sep = ' ')

    # Lista de preposiciones, artículos y stopwords en español
    stop_words_es = set(stopwords.words('spanish'))
    preposiciones_articulos = set(['a', 'ante', 'bajo', 'cabe', 'con', 'contra', 'de', 'desde', 'durante', 'en', 'entre', 'hacia', 'hasta', 'mediante', 'para', 'por', 'según', 'sin', 'so', 'sobre', 'tras', 'el', 'la', 'los', 'las', 'un', 'una', 'unos', 'unas', 'lo'])

    # Eliminación de preposiciones, artículos, stopwords y tokens con una longitud < 2
    nuevo_texto = [token for token in nuevo_texto if token not in preposiciones_articulos and token not in stop_words_es and len(token) > 1]
    
    return nuevo_texto

# Ejemplo de uso
test = "La limpieza de un texto es fundamental en la preparación de datos para el análisis"
print(test)
print(clean_tokenize(texto=test))
```

```python
df['texto_tokenizado'] = df['Texto'].apply(lambda x: clean_tokenize(x))
df[['Texto', 'texto_tokenizado']].head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.spatial.distance import cosine

# Definir la función de similitud de coseno
def similitud_coseno(a, b):
    distancia = cosine(a, b)
    return 1 - distancia

# Suponiendo que news_pivot es un DataFrame ya definido y preparado
# Calcular la matriz de correlación usando la similitud de coseno
correlation_matrix = news_pivot.corr(method=similitud_coseno)

# Crear una gráfica de calor para visualizar la matriz de correlación
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

# Agregar títulos y etiquetas
plt.title('Cosine Similarity Correlation Matrix')
plt.xlabel('Num')
plt.ylabel('Num')

# Mostrar la gráfica
plt.show()
```
